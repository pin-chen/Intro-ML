```python
import numpy as np
import pandas as pd

import pickle

from sklearn import linear_model
from sklearn.linear_model import HuberRegressor
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

import warnings
warnings.filterwarnings('ignore')
```

# Load test data, model, and sample submisson

```python
np_path = '/kaggle/input/ml-final-model/x_test.npy'
model_path = '/kaggle/input/ml-final-model/final_model.pickle'
sample_submission_path = '/kaggle/input/tabular-playground-series-aug-2022/sample_submission.csv'
```

```python
submission = pd.read_csv(sample_submission_path)
```

```python
l_x_test = np.load(np_path)
l_model = pickle.load(open(model_path, 'rb'))
```

# Predict

```python
lr_test = np.zeros(len(l_x_test))
lr_test += l_model.predict_proba(l_x_test)[:, 1]
```

# Save answer
